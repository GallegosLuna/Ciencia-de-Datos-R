*Insertamos el precio de la accion*

```r
AccionPrecio=82200
```

```r
#Empezaremos declarando las variables
Activoscirculantes=as.numeric(Base1[2,2])
Pasivoscirculantes=as.numeric(Base1[22,2])
Inventario=as.numeric(Base1[9,2])
Inventario2=Inventario*-1
Op1=as.numeric(Base1[4,2])
Op2=as.numeric(Base1[5,2])
c=c(Op1,Op2)
c=na.omit(c)
Efectivo=sum(c)
Activostotales=as.numeric(Base1[12,2])
CapitalTotal=as.numeric(Base1[36,2])
UAII=as.numeric(Base2[14,2])
Intereses=(as.numeric(Base2[15,2]))
Depreciacionyamortizacion=as.numeric(Base2[10,2])
Costovienesvendidos=as.numeric(Base2[5,2])
Ventas=as.numeric(Base2[3,2])
Cuentasporcobrar=as.numeric(Base1[7,2])
UtilidadNeta=as.numeric(Base2[26,2])
Acciones=as.numeric(Base1[47,2])
```

```r
#Medidas de liquide o solvencia a corto plazo
RazonCirculante=Activoscirculantes/Pasivoscirculantes
RP=c(Activoscirculantes, Inventario2)
RP=na.omit(RP)
RP=sum(RP)
RazonRapida=RP/Pasivoscirculantes
RazondeEfectivo=Efectivo/Pasivoscirculantes
capi=CapitalTotal*-1
RD=c(Activostotales,capi)
RD=na.omit(RD)
Rd=sum(RD)
RazondeudaTotal=Rd/Activostotales
MultiplicadordeCapital=Activostotales/CapitalTotal
Vecesganadointeres=UAII/Intereses
ce=c(UAII,Depreciacionyamortizacion)
ce=na.omit(ce)
ce=sum(ce)
CoberturaEfectivo=ce/Intereses
CoberturaEfectivo
#medidas de administracion o rotacion de activos
Rotaciondeinventario=Costovienesvendidos/Inventario
Diasdeventasenelinventario=365/Rotaciondeinventario
RotacionCuentasxcobrar=Ventas/Cuentasporcobrar
DiasdeventasCuentasporcobrar=365/RotacionCuentasxcobrar
RotaciondeActivosTotales=Ventas/Activostotales


#medidas de rentabilidad
MargenUtilidad=UtilidadNeta/Ventas
Rendimientosobreactivos=UtilidadNeta/Activostotales
Rendimientosobrecapital=UtilidadNeta/CapitalTotal

#Medidas de valor del mercado
utilidadporaccion=UtilidadNeta/Acciones
Razonprecioutilidades=AccionPrecio/utilidadporaccion
Valorenlibros=CapitalTotal/Acciones
Valordemercadoalibros=AccionPrecio/Valorenlibros
Capitalizacionmercado=AccionPrecio*Acciones

nombres1=c('razon circulante','Razon rapida','razon de efectivo', 'Razon de deuda total', 'Multiplicador del capital' 
          , 'Razon de las veces que ganamos interes', 'Razon cobertura de efectivo')
nombres2=c(RazonCirculante,RazonRapida,RazondeEfectivo,RazondeudaTotal,MultiplicadordeCapital,Vecesganadointeres,
       CoberturaEfectivo    )
Solvencia <- data.frame(nombres1,nombres2)
names(Solvencia) <- c('Razon','Medidas de solvencia')

Razones_de_administracion=c('Rotacion del inventario', 'Dias de ventas en el inventario',
                          'Razon de las cuentas por cobrar', 'Dias de ventas en cuentas por cobrar',
                          'Rotacion de los activos totales')
Medidas_de_administracion=c(Rotaciondeinventario,Diasdeventasenelinventario,RotacionCuentasxcobrar
                          ,DiasdeventasCuentasporcobrar,RotaciondeActivosTotales)

Administracion=data.frame(Razones_de_administracion,Medidas_de_administracion)
Razones_rentabilidad=c('Margen de utilidad','Rendimientos sobre activos',
                      'Rendimientos sobre capital')
Medidas_rentabilidad=c(MargenUtilidad,Rendimientosobreactivos,Rendimientosobrecapital)

rentabilidad=data.frame(Razones_rentabilidad,Medidas_rentabilidad)

Razones_valor_mercado=c('UPA','Razon precio utilidades','Razon valor de mercado a libros',
                        'Capitalizacion de mercado')
Medidas_valor_mercado=c(utilidadporaccion,Razonprecioutilidades,Valordemercadoalibros,
                        Capitalizacionmercado)
Valormercado=data.frame(Razones_valor_mercado,Medidas_valor_mercado)
```

```r
#A continuacion se generaran data frames con las razones separadas por "rubro".
Solvencia
Administracion
rentabilidad
Valormercado
```

## Conclusiones obtenidas con las medidas de liquidez

```r

if(RazonCirculante<1){Print('La empresa no puede solventar obligaciones a corto plazo')}else{print('La empresa puede cumplir sus obligaciones a corto plazo')}
if(RazonRapida<1){print('La empresa no puede convertir rapidamente sus activos mas liquidos en efectivo al corto plazo')}else{print('La empresa convertir en efectivo rapidamente sus activos mas liquidos en el corto plazo')}
if(RazondeEfectivo<1){print('No posee los suficientes fondos para cubrir sus deudas al corto plazo')}else{print('Posee los fondos suficientes para cubrir la deuda en el corto plazo')}
if(MultiplicadordeCapital==1){print('Se ha financiado con recursos propios')}else{print('Obtuvo ayuda de terceros para financiar sus proyectos')}
```

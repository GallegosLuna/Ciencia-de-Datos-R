## KIA

```r
library(quantmod)
library(knitr)
getSymbols("000270.KS")#extraemos los datos financiero de KIA
KIA <- xts::last(`000270.KS`,'2 years')#Extraeremos la informacion de los ultimos24 meses
```

## Graficaremos una serie de tiempo con velas

```r
chartSeries(`000270.KS`, name='Serie de tiempo KIA',subset='2020-01::2022-04',TA=NULL)
```

## Realizamos el mismo grfico pero con la grafica de volumen

```r
chartSeries(`000270.KS`, name="Serie de tiempo KIA",
            
            subset='2020-01::2022-04',
            theme=chartTheme('white'))
```

## Tambien podemos personalizar nuestros graficos

```r
#Personalizamos los graficos
# Tema personalizado (lo ponemos en modo de funcion)
myTheme <- chart_theme()
myTheme$col$dn.col <- "pink"
myTheme$col$dn.border <- "pink"
myTheme$col$up.col <- "green"
myTheme$col$up.border <- "green"
myTheme$rylab <- TRUE
myTheme$lylab <- FALSE

# Dibujamos la serie
chart_Series(`000270.KS`, theme = myTheme, name='Serie de tiempo KIA',subset='2020-01::2022-04')
```

## Preferentemente usamos este tipo de graficas con algunos indicadores.

## En este caso usare medias moviles, con el fin de crear estrategias

```r
# Dibujamos la serie
chart_Series(`000270.KS`,
             TA = list("add_EMA(n = 20, col = 4,
                                lwd = 2)",
                       "add_EMA(n = 50, col = 2,
                                lwd = 2)"),subset='2020-01::2022-04',name='Serie de tiempo KIA')
```
